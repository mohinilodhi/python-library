## welcome to the matplotlib library practice


```python
import matplotlib.pyplot as plt

#simple data
marks = [80, 90, 76, 55, 95]
students = ["abhay", "aakansha","akash","ajay","aarya"]
plt.plot(marks)
plt.title("marks Line Graph")
plt.show()
```

```python
plt.bar(students, marks)
plt.title("student Marks")
plt.show()
```

```python
plt.hist(marks)
plt.title("marks histogram")
plt.show()
```

```python
age = [18, 19, 20, 18, 19]
plt.scatter(age, marks)
plt.title("age vs marks")
plt.show()
```

```python
plt.pie(marks)
plt.title("marks pie chart")
plt.show()
```

```python
plt.xlabel(marks)
plt.title("marks distribution")
plt.show()
```

```python
plt.ylabel(students)
plt.title("students")
plt.show()
```

matplotlib library with pandas





```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "name": ["a", "b", "c", "d", "e", "f"],
    "subject": ["maths", "sst", "science", "hindi", "ai", "ep"],
    "marks": [12, 13, 14, 15, 16, 17],
    "age": [18, 19, 20, 21, 22, 23]  # Added a sample 'age' column
}
df = pd.DataFrame(data)
print(df)
```

```python
df["marks"].plot()
plt.title("marks line graph")
plt.show
```

```python
df.plot(x="name", y="marks", kind="bar")
plt.title("name vs marks")
plt.show()
```

```python
df["marks"].plot(kind="hist")
plt.title("marks histogram")
plt.show()
```

```python
df.plot()
plt.show()
```

```python
df.plot(x='age', y='marks', kind='scatter')
plt.title('Age vs Marks from DataFrame')
plt.show()
```

```python
df["studyHours"] = [2, 3, 4, 5, 6, 7]

df.plot(x="studyHours", y="marks", kind="scatter")
plt.show()
```
