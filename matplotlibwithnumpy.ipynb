```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4,5,6,7,8])
y = np.array([3, 4, 5, 6,7,8,9,10])

plt.bar(x,y)
plt.title("bar graph using numpy")
plt.show
```

```python
import numpy as np
import matplotlib.pyplot as plt

marks = np.array([50,40, 60, 10])

plt.pie(marks, labels =["A","B","C","D"], autopct="%1.1f%%")
plt.title("marks distribution")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.array([1,2,3,4])

y1=np.array([40, 50,60,70,])
y2=np.array([30, 40,50,60])

plt.plot(x, y1, label=" Line 1")
plt.plot(x, y2, label=" Line 2")

plt.legend()
plt.title("multiple lines")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.subplot(1,2,1)
plt.plot([1,2,3],[10,20,30])
plt.title("subplot 1")

plt.subplot(1,2,2)
plt.plot([1,2,3],[10,20,30])
plt.title("subplot 2")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.array([1,2,3,4,5])
y=np.array([10,20,30,40,50])

plt.fill_between(x,y,color="red")
plt.title("fill between")
plt.show()
```
